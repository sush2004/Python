```
sns.stripplot(y= 'tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```
sns.stripplot(x='day', y='tip', data=tips, size=6,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```
# set jitter = false for thin strips
sns.stripplot(x='day', y='tip', data=tips, size=6,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```
sns.swarmplot(x='day', y='tip',size=3, data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient = 'h')
plt.ylabel('tip ($)')
plt.show()
```

```
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```
